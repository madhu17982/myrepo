# Finding Datatypes

```python
def find_datatype():
    numCols = data.select_dtypes("number").columns
    catCols = data.select_dtypes("object").columns
    print("Numerical Data  ")
    print("---------------------------------------")
    numCols= list(set(numCols))
    print(numCols)
    
    print("\nCatagorical Data ")
    catCols= list(set(catCols))
    print("---------------------------------------")
    print(catCols)
find_datatype()
```

```python
df =pd.DataFrame(data,columns=['Prev Close','Open'])
df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

# Spliting Data in to Training and Test sets For Analysis

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
lin_pred=lin_reg.predict(X_test)
```

```python
# Fitting Support vector Regression to the dataset
from sklearn.svm import SVR
sv_regressor = SVR(kernel = 'linear')
sv_regressor.fit(X_train, y_train)
svr_pred=sv_regressor.predict(X_test)
```

```python
# Fitting KNeighbors Regression to the dataset
from sklearn.neighbors import KNeighborsRegressor
knn_regressor=KNeighborsRegressor(n_neighbors = 5)
knn_regressor=knn_regressor.fit(X_train,y_train)
knn_pred=knn_regressor.predict(X_test)
```

# Evaluating Regression Models

# Cross Validation

```python
from sklearn import model_selection
from sklearn.model_selection import KFold
kfold = model_selection.KFold(n_splits=20, random_state=100,shuffle=True)
linear_kfold = model_selection.cross_val_score(lin_reg, X_test, y_test.astype('int'), cv=kfold)
svr_kfold = model_selection.cross_val_score(sv_regressor, X_test, y_test.astype('int'), cv=kfold)
knn_kfold = model_selection.cross_val_score(knn_regressor, X_test, y_test.astype('int'), cv=kfold)
print("LinearRegression Accuracy: ", linear_kfold.mean()*100)

print("-------------------------------------------------------------------------------------")
print("SupportVector_Regression Accuracy: ", svr_kfold.mean()*100)
print("-------------------------------------------------------------------------------------")
print("KNeighbors_Regressor Accuracy: ", knn_kfold.mean()*100)
```

# Mean squared error 

```python
from sklearn.metrics import mean_squared_error 
import math
lr_mse=math.sqrt(mean_squared_error(y_test,lin_pred))
print('Linear Model Root mean square error',lr_mse)
svr_mse=math.sqrt(mean_squared_error(y_test,svr_pred))
print('Support vector model Root mean square error',svr_mse)
knn_mse=math.sqrt(mean_squared_error(y_test,knn_pred))
print('KNN Model Root mean square error',knn_mse)
```

```python
from sklearn.metrics import r2_score
print('Linear R2: ', r2_score(y_test, lin_pred))
print('SVR R2: ', r2_score(y_test, svr_pred))
print('Knn R2: ', r2_score(y_test, knn_pred))
```
